This notebook is to get the links of the QNB press releases and arrange them yearwise

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

URL = "https://www.qnb.com/sites/qnb/qnbglobal/page/en/ennews.html"
page = requests.get(URL)

print(page.text)
```

```python
soup = BeautifulSoup(page.content, "html.parser")
```

```python
year_wise_news_elements = soup.find_all("div", class_="acc")
```

```python
# find the maximum number of articles for any year
max_num_articles_per_year=-1

for year_wise_news_element in year_wise_news_elements:
    year=year_wise_news_element.find_all('h2')[0].text
    list_this_year=year_wise_news_element.find_all('ul',class_='table-date-group newsroom_listing')[0]
    list_elements=list_this_year.find_all('li')
    if max_num_articles_per_year<len(list_elements):
        max_num_articles_per_year=len(list_elements)
```

```python
max_num_articles_per_year
```

```python
# thus we need a dataframe
```

```python
dic_data={}
for year_wise_news_element in year_wise_news_elements:
    year=year_wise_news_element.find_all('h2')[0].text
    if year not in dic_data:
        dic_data[year+"_title"]=[]
        dic_data[year+"_url"]=[]        
        dic_data[year+"_date"]=[]                
    print(year)
    list_this_year=year_wise_news_element.find_all('ul',class_='table-date-group newsroom_listing')[0]
    list_elements=list_this_year.find_all('li')
    print(len(list_elements))
    for elem in list_elements:
        title=elem.find_all('div')[0].text
        url=elem.find_all('a')[0]["href"]
        date=elem.find_all('div',class_='date')[0].text
        dic_data[year+"_title"].append(title)
        dic_data[year+"_url"].append(url)
        dic_data[year+"_date"].append(date)
```

```python
dic_data
```

```python
for k,v in dic_data.items():    
    if len(dic_data[k])<max_num_articles_per_year:
        extra=[' ' for i in range(max_num_articles_per_year-len(dic_data[k]))]
        dic_data[k].extend(extra)
        
```

```python
df_dic=pd.DataFrame(dic_data)
```
