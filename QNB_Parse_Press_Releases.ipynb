```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def get_freq_from_list_words(list_words,list_keywords):
    '''
    returns a dictionary with key being each search term
    and the number of times the key was present
    '''
    dic_keywords={}
    for keyword in list_keywords:
        dic_keywords[keyword]=0
    for i in range(len(list_words)):
        word=list_words[i]
        for keyword in list_keywords:
            if keyword in word or word in keyword:
                dic_keywords[keyword]+=1
    return dic_keywords
```

```python
years=[col.split("_")[0] for col in list(df.columns)]
years=set(years)


base_URL="http://www.qnb.com"
dataframe_stats={}
dataframe_stats["year"]=[]
dataframe_stats["date"]=[]
dataframe_stats["title"]=[]
dataframe_stats["URL"]=[]
for kw in keywords:
    dataframe_stats[kw]=[]
    
```

```python


    
for year in years:
    if int(year)!=2022:
        continue

    titles=list(df[year+"_title"])
    urls=list(df[year+"_url"])
    dates=list(df[year+"_date"])    
    print(year)    
    
    for i in range(70,len(titles)):
        if i%10==0:
            print("\t",i,"/",len(titles))
        title=titles[i]
        if title==' ':
            break
        if urls[i]==' ':
            continue
        full_url=base_URL+urls[i]
        date=dates[i]
        page = requests.get(full_url)
        
        soup = BeautifulSoup(page.content, "html.parser")
        subpage_content = soup.find_all("div", class_="page-subpage-content")

        if len(subpage_content)>0:
            page_text=subpage_content[0].text
            page_text=page_text.replace("\n","")
            if "\xa0" in page_text:
                page_text=page_text.replace("\xa0","")    
                
            page_text=page_text.lower()
            words=page_text.split()
            dic_kw=get_freq_from_list_words(words,keywords)
            dataframe_stats["year"].append(year)
            dataframe_stats["date"].append(date)
            dataframe_stats["title"].append(title)
            dataframe_stats["URL"].append(full_url)            
            for kw in keywords:
                dataframe_stats[kw].append(dic_kw[kw])            
            
```
